```python
BHolidayNo = 0.3
BYesDiscountNo = 0.4
BYesDeliveryNo = 0.5
BHoliday = 0.5
BYesDiscount = 0.6
BYesDelivery = 0.7
BNo = 0.4
```

```python
BNoBuy1 = (BHolidayNo * BYesDiscountNo * BYesDeliveryNo) * BNo / (BHoliday * BYesDiscount * BYesDelivery)
print(BNoBuy1)
```

```python
# Kondisi B dimana
# Day Holiday
#Discount = Yes
#Free Delivery = Yes
# P(A)= Buy = BYes
#P(B)= BYesDiscount, ByesDelivery, BHoliday
#P(B|A)1 = (BHoliday|Buy) = BHolidayYes
#P(B|A)2 = (BYesDiscountYes|Buy) = BYesDiscountYes
#P(B/A)3 = (BYesDeliveryYes|Buy) = BYesDeliveryYes
# P(B)1 = BHoliday
#P(B)2 = BYesDiscount
#P(B)3 = BYesDelivery

#ditanyakan P(A/B) ?
#Rumus
#P(AB) P(B|A)*B(A)/P(B)
```

```python
BYes = 0.6     # contoh nilai probabilitas 'Buy = Yes'
BHolidayYes = 0.7
BYesDiscountYes = 0.8
BYesDeliveryYes = 0.9
BHoliday = 0.5
BYesDiscount = 0.6
BYesDelivery = 0.7
```

```python
BBuy2 = (BHolidayYes * BYesDiscountYes * BYesDeliveryYes) * BYes / (BHoliday * BYesDiscount * BYesDelivery)
print(BBuy2)
```
